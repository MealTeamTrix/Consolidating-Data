```python
# Import libraries
import pandas as pd
```

```python
# Load data
offices = pd.read_csv("data/office_addresses.csv")
addresses = pd.read_excel("data/employee_information.xlsx")
```

```python
# Define header for emergency contacts
emergency_contacts_cols = [
    "employee_id", "last_name", "first_name",
    "emergency_contact", "emergency_contact_number", "relationship"
]

# Load emergency contacts sheet
emergency_contacts = pd.read_excel(
    "data/employee_information.xlsx",
    sheet_name="emergency_contacts",
    header=None,
    names=emergency_contacts_cols
)
```

```python
# Load roles
roles = pd.read_json("data/employee_roles.json", orient="index")
```

```python
# Merge all data sources
employees = (
    addresses
    .merge(offices, how="left", left_on="employee_country", right_on="office_country")
    .merge(roles, how="left", left_on="employee_id", right_index=True)
    .merge(emergency_contacts, how="left", on="employee_id")
)
```

```python
# Fill missing office information
office_columns = ["office", "office_country", "office_city", "office_street", "office_street_number"]
employees[office_columns] = employees[office_columns].fillna("Remote")
```

```python
# Select final columns
final_columns = [
    "employee_id", "employee_first_name", "employee_last_name", "employee_country", 
    "employee_city", "employee_street", "employee_street_number", 
    "emergency_contact", "emergency_contact_number", "relationship", 
    "monthly_salary", "team", "title", "office", "office_country", 
    "office_city", "office_street", "office_street_number"
]
```

```python
employees_final = employees[final_columns].set_index("employee_id")

# View the results
employees_final
```
